# Clustering

```python
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
```

```python
datacl=data.copy()
datacl.sample()
```

```python
datacl=datacl[['hour','days', 'avg_delay']]
```

```python
cat_var = ['hour','days']
for var in cat_var:
    cat_list = pd.get_dummies(datacl[var], prefix=var)
    data1 = datacl.join(cat_list)
    datacl = data1
datacl.sample()
```

```python
datacl.sample()
```

```python
datacl=datacl.drop(columns=['hour', 'days'])
datacl.sample()
```

```python
x_array = np.array(datacl)
print(x_array)
```

```python
scaler = MinMaxScaler()
x_scaled = scaler.fit_transform(x_array)
x_scaled
```

```python
dbi = []
K = range(2,10)
from sklearn.cluster import KMeans
from sklearn.metrics import davies_bouldin_score
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(x_scaled)
    labels=km.labels_
    dbi.append(davies_bouldin_score(x_scaled, labels))
```

```python
plt.style.use("fivethirtyeight")
plt.plot(range(2, 10), dbi)
plt.xticks(range(2, 10))
plt.xlabel("Number of Clusters")
plt.ylabel("DBI")
plt.show()
```

```python
dbi
```

```python
kmeans=KMeans(n_clusters=3)
kmeans.fit(x_scaled)
```

```python
klaster=pd.DataFrame(kmeans.labels_)
```

```python
datacl['klaster']=klaster
```
